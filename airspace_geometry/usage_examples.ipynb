# Setup

```python
from airspace_geometry import (
    circle,
    circular_sector,
    circular_segment,
    ring,
    ring_sector
)
from _types import GeographicCoordinates
```

# Airspace geometries

```python
circle_ = circle(center=GeographicCoordinates(20, 50),
                 radius=5000)
circle_
```

```python
circular_sector_ = circular_sector(center=GeographicCoordinates(20, 50),
                                   radius=5000,
                                   azimuth_from=10.5,
                                   azimuth_to=37.5)
circular_sector_
```

```python
ring_ = ring(center=GeographicCoordinates(20, 50),
             inner_radius=2400,
             outer_radius=5000)
ring_
```

```python
circular_segment_ = circular_segment(center=GeographicCoordinates(20, 50),
                                     azimuth_from=10,
                                     azimuth_to=120,
                                     radius=4500)
circular_segment_
```

```python
ring_sector_ = ring_sector(center=GeographicCoordinates(20, 50),
                           inner_radius=1000,
                           outer_radius=4000,
                           azimuth_from=330.5,
                           azimuth_to=47.5)
ring_sector_
```
